```python
# your code here
#defining a function to generate normally distributed random numbers 

def generate_normal_random(mean,std,count):
    return np.random.normal(mean,std,count)
```

Which one(s) of the variables are nearly normally distributed? How do you know?

```python
# your code here
#to confirm the plots distribution, the skewness should be calculated 
from scipy.stats import skew

def calculated_skewness(df,columns):
    skewness_dict={}
    for column in columns:
        skewness_dict[column] = skew(df[column])
    return skewness_dict
```

```python
# your code here

#defining the columns we want to calculate the skewness
columns_of_interest = ['Fuel Barrels/Year', 'CO2 Emission Grams/Mile','Combined MPG']

#Calling the functions
skewness_results = calculated_skewness(vehicles_df,columns_of_interest)

#printing the results 
for column,skewness in skewness_results.items():
    print(f'Skeness of {column}:{skewness}')
```

```python
# your code here
# In this case, none of the variables above are normally distributed, they all tilt to the right. Being Fuel Barrels/Year the one closes to 0 but hard to defined as 'normally distributed'
```
